### **Background and Introduction**

```python
#defining function for checking missing values 
#(The following function returns a summary table showing the count and percentage of missing values in each column)

def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()
    
    # Percentage of missing values
    mis_val_percent = 100 * mis_val / len(df)
    
    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    
    # Rename the columns explicitly
    mis_val_table.columns = ['Missing Values', '% of Total Values']
    
    # Filter out columns with no missing data and sort descending
    mis_val_table = mis_val_table[mis_val_table['Missing Values'] != 0]
    mis_val_table = mis_val_table.sort_values('% of Total Values', ascending=False).round(1)
    
    # Print summary
    print(f"Selected dataframe has {df.shape[1]} columns.")
    print(f"There are {mis_val_table.shape[0]} columns that have missing values.")
    
    return mis_val_table
```

AQI Bucket is defined by the actual AQI value. Hence, it will be prudent to drop 'AQI_Bucket' and replace it with 'AQI_Bucket_Calculated' to ensure data consistency. For calculation, we will use the parameters as laid down by CPCB to define a function to calculate AQI Bucket based on the AQI value. Laid down categorization can be accessed here https://airquality.cpcb.gov.in/AQI_India/ 

```python
#defining AQI categorization function
def categorize_aqi(AQI):
    if AQI <= 50:
        return "Good"
    elif AQI <= 100:
        return "Satisfactory"
    elif AQI <= 200:
        return "Moderate"
    elif AQI <= 300:
        return "Poor"
    elif AQI <= 400:
        return "Very Poor"
    else:
        return "Severe"
```

```python
#creating a copy
df_interpolated_clean = df_interpolated_clean.copy()

#applying categorization to create new column
df_interpolated_clean['AQI_Bucket_Calculated'] = df_interpolated_clean['AQI'].apply(categorize_aqi)
```

## **Modeling Phase**

# **Part-2**

In Part-1 of the project, we noticed near-perfect results across Decision Tree and Random Forest, plus ~0.98 from Logistic Regression, strongly suggesting **target leakage**.

The feature set still includes `AQI`, while the target `AQI_Bucket` is defined from `AQI`. In this setup even a shallow tree can split on `AQI` thresholds and exactly reproduce the bucket, so cross-validation also looks perfect (the leak exists in every fold). A tell-tale symptom you should see is AQI ranking overwhelmingly #1 in feature importance and a perfectly diagonal confusion matrix.

*Verification step*. On the current (leaky) model, computing feature importance will clearly confirm leakage, if AQI dominates by a wide margin.

*Remediation*. Remove AQI (and any columns derived from it) from the feature matrix and re-run the entire modeling block (split, CV, tuning, confusion matrices, normalized CM, per-class metrics, and permutation importance). Metrics will drop to realistic values, and the resulting importances will reflect true signal from pollutants and weather rather than the leaked target. To prevent recurrence, keep a single guard in code (e.g., assert that no column name contains “AQI”) before training.

```python
#dropping AQI column
df_model_noAQI = df_clean.drop(columns=['AQI'])
df_model_noAQI.info()
```

```python
# ==== Imports ====
import numpy as np
import pandas as pd
import os
import joblib

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.base import clone
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, f1_score, cohen_kappa_score, ConfusionMatrixDisplay,
    make_scorer, classification_report, confusion_matrix, precision_recall_fscore_support
)
from sklearn.model_selection import (
    train_test_split, StratifiedKFold, cross_validate, RandomizedSearchCV, GridSearchCV
)
from sklearn.tree import DecisionTreeClassifier

from xgboost import XGBClassifier
```

```python
# ==== Global config ====
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# ==== Cross-validation + metrics ====
cv3 = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE)

SCORING = {
    "accuracy": "accuracy",
    "macro_f1": make_scorer(f1_score, average="macro"),
    "kappa": make_scorer(cohen_kappa_score),
}
PRIMARY_SCORE = "macro_f1"   # used as refit metric in searches
```

### **Encoding, target, split and class names**

```python
# Encoding target
le2 = LabelEncoder()
y2 = le2.fit_transform(df_model_noAQI['AQI_Bucket'])
class_names2 = le2.classes_

# Feature / target
X2 = df_model_noAQI.drop(columns=['AQI_Bucket'])

# Split (stratified)
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X2, y2, test_size=0.20, stratify=y2, random_state=RANDOM_STATE
)

print("Train:", X_train2.shape, "| Test:", X_test2.shape)

# Pretty class distribution
dist_train = pd.Series(y_train2).value_counts(normalize=True).sort_index()
dist_test  = pd.Series(y_test2).value_counts(normalize=True).sort_index()
name_map   = dict(enumerate(class_names2))

print("\nTrain dist:\n", dist_train.rename(index=name_map).round(4))
print("\nTest  dist:\n", dist_test.rename(index=name_map).round(4))
```

```python
# Carrying out human readable mapping of names
code_to_name = dict(enumerate(class_names2))
print(code_to_name)  # {0: 'Good', 1: 'Moderate', ...}

# For printing distributions
print("\nTrain dist (named):\n", dist_train.rename(index=code_to_name).round(4))
print("\nTest  dist (named):\n",  dist_test.rename(index=code_to_name).round(4))

# Ensuring consistent label order in confusion matrices
cm_labels = list(class_names2)  # pass as xticklabels/yticklabels
```

```python
# Defining functions

def plot_cm(y_true, y_pred, labels, title):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
                xticklabels=labels, yticklabels=labels)
    plt.title(title); plt.xlabel("Predicted"); plt.ylabel("Actual")
    plt.xticks(rotation=45); plt.tight_layout(); plt.show()

def test_row(name, y_true, y_pred):
    return {
        "Model": name,
        "Accuracy": round(accuracy_score(y_true, y_pred), 4),
        "Macro F1": round(f1_score(y_true, y_pred, average="macro"), 4),
        "Kappa": round(cohen_kappa_score(y_true, y_pred), 4),
    }

def tune_then_cv_single(name, est, params, *, X=X_train2, y=y_train2, n_iter=12, cv=cv3, verbose=1):
    rs = RandomizedSearchCV(
        estimator=est,
        param_distributions=params,
        n_iter=n_iter,
        scoring=SCORING, refit=PRIMARY_SCORE,
        cv=cv, n_jobs=-1, random_state=RANDOM_STATE, verbose=verbose
    )
    rs.fit(X, y)
    best_est = rs.best_estimator_
    cv_res = cross_validate(best_est, X, y, scoring=SCORING, cv=cv, n_jobs=-1)
    row = {
        "Model": f"{name} (tuned)",
        "Accuracy (CV)": round(np.mean(cv_res["test_accuracy"]), 4),
        "Macro F1 (CV)": round(np.mean(cv_res["test_macro_f1"]), 4),
        "Kappa (CV)": round(np.mean(cv_res["test_kappa"]), 4),
        "Best Params": rs.best_params_,
    }
    return {"row": row, "best_estimator": best_est, "search": rs}
```

### **Baselines (Dummy, LR pipeline) with CV**

```python
# Quadratic-weighted kappa scorer

SCORING["qwk"] = make_scorer(cohen_kappa_score, weights="quadratic")

# Baselines (keep plain + weighted)
dummy = Pipeline([("clf", DummyClassifier(strategy="most_frequent"))])

# Pipeline ensures scaler fit is fold-local (prevents CV leakage)
lr_base = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=1000, random_state=RANDOM_STATE, multi_class="auto"))
])

lr_bal = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=1000, random_state=RANDOM_STATE, class_weight="balanced", multi_class="auto"))
])

def cv_row(name, estimator, X=X_train2, y=y_train2):
    cv = cross_validate(estimator, X, y, scoring=SCORING, cv=cv3, n_jobs=-1)
    return {
        "Model": name,
        "Accuracy (CV)": round(cv["test_accuracy"].mean(), 4),
        "Macro F1 (CV)": round(cv["test_macro_f1"].mean(), 4),
        "Kappa (CV)":     round(cv["test_kappa"].mean(), 4),
        "QWK (CV)":       round(cv["test_qwk"].mean(), 4),
        "Best Params": None,
    }

baseline_cv_df = pd.DataFrame([
    cv_row("Dummy (Most Frequent)", dummy),
    cv_row("Logistic Regression (baseline)", lr_base),
    cv_row("Logistic Regression (balanced)", lr_bal),
]).sort_values(["Macro F1 (CV)", "Kappa (CV)"], ascending=False).reset_index(drop=True)

baseline_cv_df
```

   - Dummy Classifier: As expected, the “most frequent” baseline performs very poorly (Accuracy ≈ 0.26, Macro-F1 ≈ 0.07, Kappa = 0). This confirms the dataset is imbalanced and highlights the need for models that can capture more than majority-class prediction.
   
   - Logistic Regression (standard): Shows a clear improvement, achieving ~62% accuracy and Macro-F1 ~0.64. Importantly, the Quadratic Weighted Kappa (QWK) = 0.66 indicates that predictions respect the ordinal nature of AQI categories (closer agreement with actual labels than raw accuracy).
  
   - Logistic Regression (class-balanced): Slightly lower overall performance compared to unweighted LR. While balancing helped reduce bias toward majority classes, it did not improve Macro-F1 or QWK, suggesting that the distribution imbalance is not the only challenge—category overlap (e.g., Moderate vs. Poor) is also significant.

3. *Observations*

```python
# Tuning Models

summary_cv = pd.DataFrame([])
res_store  = {}


# ---- Random Forest ----
rf = RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1)
rf_params = {
    "n_estimators": [100, 150, 200, 300],
    "max_depth": [None, 8, 12, 16],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"],
}
rf_out = tune_then_cv_single("Random Forest", rf, rf_params, n_iter=12, verbose=1)
summary_cv = pd.concat([summary_cv, pd.DataFrame([rf_out["row"]])], ignore_index=True)
res_store["Random Forest"] = rf_out
```

- Rare categories like Good and Satisfactory show slightly lower recall, but they represent a very small proportion of the dataset.

**Interpretation**
- The model captures fine-grained distinctions across AQI categories, showing robustness against imbalance and delivering meaningful public health predictions.

**Next Action**

- We will now carry out GridSearchCV on the Best Model (XGB).

```python
# Performing GridSearchCV on the Best Model (XGB)

xgb_grid = {
    "n_estimators": [best_est.get_params()["n_estimators"]-50, best_est.get_params()["n_estimators"], 
                     best_est.get_params()["n_estimators"]+50],
    "max_depth":    [max(3, best_est.get_params()["max_depth"]-1), best_est.get_params()["max_depth"], 
                     best_est.get_params()["max_depth"]+1],
    "learning_rate":[max(0.01, best_est.get_params()["learning_rate"]/2), best_est.get_params()["learning_rate"], 
                     min(0.3, best_est.get_params()["learning_rate"]*1.5)]
}
gs = GridSearchCV(best_est, xgb_grid, scoring=SCORING, refit="macro_f1", cv=cv3, n_jobs=-1)
gs.fit(X_train2, y_train2)
print("Grid best params:", gs.best_params_)
```

     - n_estimators = 400
